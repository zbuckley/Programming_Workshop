## Basic Matplotlib graphs

```python
import matplotlib.pyplot as plt
import numpy as np
plt.plot([1, 2, 3, 4])
plt.ylabel('some numbers')
plt.show()
#when we provide only a single list, it interprets it as y axis values and calculates values for x axis.
```

```python
#We can provide both x axis and y aixs
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.show()
```

```python
#We can change the format the style of the plot
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro')
plt.axis([0, 6, 0, 20]) # this has [xmin,xmax,ymin,ymax]
plt.show()
```

```python
#Creating Scatter Plot
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

plt.scatter('a', 'b', c='c', s='d', data=data) #c is for color, s is for size
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

References:- https://matplotlib.org/tutorials/introductory/pyplot.html

```python
#Using bar graphs and multiple graphs 
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]

#plt.plot([1,3])
plt.figure(1,figsize=(13,4))

plt.subplot(131)
plt.bar(names, values)
plt.subplot(132)
plt.scatter(names, values)
plt.subplot(133)
plt.plot(names, values)
plt.suptitle('Categorical Plotting')
plt.show()
```
